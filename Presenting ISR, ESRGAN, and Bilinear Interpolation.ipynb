```python
# https://pyimagesearch.com/2014/09/15/python-compare-two-images/
def mse(imageA, imageB):
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    return err

def compare_images(imageA, imageB, title):
    m = mse(imageA, imageB)
    s = ssim(imageA, imageB)

    fig = plt.figure(title)
    plt.suptitle("MSE: %.2f, SSIM: %.2f" % (m, s))

    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(imageA, cmap = plt.cm.gray)
    plt.axis("off")

    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(imageB, cmap = plt.cm.gray)
    plt.axis("off")

    plt.show()
```

```python
from sklearn import preprocessing

# Diff images
ESRGAN_B = cv2.imread('Images/GOES-B/GOES-B-True ESRGAN.png')
ISR_B = cv2.imread('Images/GOES-B/GOES-B-True ISR.jpg')
UNSHARP_MASK_B = cv2.imread('Images/GOES-B/GOES-B Unsharp Mask.jpg')
BI_B = cv2.imread('Images/GOES-B/GOES-B BI.jpg')

ESRGAN_B_grey = cv2.cvtColor(ESRGAN_B, cv2.COLOR_BGR2GRAY)
ISR_B_grey = cv2.cvtColor(ISR_B, cv2.COLOR_BGR2GRAY)
UNSHARP_MASK_B_grey = cv2.cvtColor(UNSHARP_MASK_B, cv2.COLOR_BGR2GRAY)
BI_B_grey = cv2.cvtColor(BI_B, cv2.COLOR_BGR2GRAY)

diff_1 = BI_B_grey.copy()
#diff_1 = 255 + cv2.absdiff(BI_B_grey, BI_B_grey) #white
diff_1 = cv2.absdiff(BI_B_grey, BI_B_grey).astype(np.float) #black

diff_2 = BI_B_grey.copy()
#diff_2 = 255 - cv2.absdiff(BI_B_grey, ESRGAN_B_grey) #white
diff_2 = cv2.absdiff(BI_B_grey, ESRGAN_B_grey).astype(np.float) #black

diff_3 = BI_B_grey.copy()
#diff_3 = 255 - cv2.absdiff(BI_B_grey, ISR_B_grey) #white
diff_3 = cv2.absdiff(BI_B_grey, ISR_B_grey).astype(np.float) #black

diff_4 = BI_B_grey.copy()
#diff_4 = 255 - cv2.absdiff(BI_B_grey, UNSHARP_MASK_B_grey) #white
diff_4 = cv2.absdiff(BI_B_grey, UNSHARP_MASK_B_grey).astype(np.float) #black

scaler = preprocessing.MinMaxScaler()
diff_1_norm = scaler.fit_transform(diff_1)
diff_2_norm = scaler.fit_transform(diff_2)
diff_3_norm = scaler.fit_transform(diff_3)
diff_4_norm = scaler.fit_transform(diff_4)

print("Diff_1: " + str(diff_1))
print()
print("Diff_1_norm: " + str(diff_1_norm))
print()
print("Diff_2: " + str(diff_2))
print()
print("Diff_2_norm: " + str(diff_2_norm))
print()
print("Diff_3: " + str(diff_3))
print()
print("Diff_3_norm: " + str(diff_3_norm))
print()
print("Diff_4: " + str(diff_3))
print()
print("Diff_4_norm: " + str(diff_3_norm))
```
